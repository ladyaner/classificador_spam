## 0.1 Funções auxiliares

```python
#Função para remoção de emojis
#código de referência da função: https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)



#Função para preprocessamento do texto: com remoção de números, pontuação e caracteres especiais
def preprocessamento(texto):
    #removendo emojis 
    texto= remove_emojis(texto) 

    #removendo quebra de linha
    texto= re.sub(r'\n+',' ',texto)
       
    #removendo pontuações e digitos 
    texto = ''.join([str(letra) for letra in texto if not letra.isdigit() and letra not in string.punctuation])
    return texto


#Função para GridSearch 
def grid_s(model, parametros, cv, xdata, ydata):
    grids= GridSearchCV(estimator= model, param_grid= parametros,
                        scoring='accuracy',cv=cv, n_jobs=-1)
    grids.fit(xdata, ydata)
    melhores_parametros= grids.best_params_
    melhor_resultado= grids.best_score_
    return melhores_parametros, melhor_resultado


#Função para fazer a previsão se é Spam ou Não - Classificador ESpam
def ESpam(mensagem):
    mensagem= pd.Series(mensagem)
    mensagem.apply(preprocessamento)
    pred= classificador_nb.predict(mensagem)
    return pred
```

## 3.1 Seleção: variável de entrada e classe(rótulo)

```python
X_train= df_train['Message'] #Seleção da coluna de mensagens        - Variável de entrada -
Y_train= df_train['SPAM']    #seleção da classe SPAM: True e False  - Label do Classificador -
```

## 3.2 Processamento dos dados

### 3.2.1 Limpeza inicial dos dados

```python
#Limpeza inicial dos dados- aplicar a função preprocessamento
x_train = X_train.apply(preprocessamento)
```

### 3.2.2 Tokenização/ Vetores

Nessa etapa utilizaremos a função CountVectorizer do sklearn, que já faz o processo de tokenização e transforma os tokens em vetores.

```python
#Instanciando função CountVectorizer
cvt= CountVectorizer(strip_accents='ascii',           #remove acentos
                     lowercase=True,                  #letras minúsculas
                     stop_words=sw_cloud)             #remove stopwords

#transformando dados
x_train_cvt= cvt.fit_transform(x_train)
```

```python
print(x_train_cvt.toarray()) #matriz binária 
```

### 3.2.3 TF-IDF

Utilizaremos a função TfidfTransformer do pacote sklearn para normalizar a matriz gerada e com as medidas estatisticas  TF-IDF representar a importância das palavras no texto.

```python
#Intanciando função TfidfTransformer
tfi= TfidfTransformer(use_idf= True)  #parâmetro use_idf: True para normalizar

#transformação dos dados
x_train_tfi= tfi.fit_transform(x_train_cvt)
```

## 3.3 Separação dos dados de treino e teste 

```python
#separação dos dados de treino e teste- 20% para teste
xtrain, xtest, ytrain, ytest= train_test_split(x_train_tfi, Y_train, test_size= 0.2, random_state= 17)

#verificar os shapes
print(f'xtrain: {xtrain.shape} /ytrain: {ytrain.shape}')
print(f'xtest: {xtest.shape} /ytest: {ytest.shape}')
```

## 6.1  Construção do Pipeline

```python
#instanciando CountVecotizer
cvt= CountVectorizer(strip_accents='ascii', lowercase=True, stop_words= sw_cloud)

#instanciando TfidfTransformer
tfi= TfidfTransformer(use_idf= True)

#instanciando modelo NB
classificador= MultinomialNB(alpha=1.0)

#Pepiline com classificador NB
clf_pipeline= Pipeline(steps=[('CountVectorizer', cvt),
                              ('TfidfTransformer', tfi),
                              ('Modelo', classificador)])
```

## 6.2 Treinamento com pipeline 

```python
#dados de treino e teste
xtreino, xteste, ytreino, yteste= train_test_split(X_train, Y_train, test_size= 0.2,random_state=1)

#preprocessamento dos dados
xtreino= xtreino.apply(preprocessamento)
xteste= xteste.apply(preprocessamento)

#treinamento do modelo com pipeline
classificador_nb = clf_pipeline.fit(xtreino,ytreino)

#predição do modelo
y_pred= classificador_nb.predict(xteste)
```

## 6.3 Salvando Moldelo Treinado

```python
#modelo treinado com toda base de dados de treino
xtreino= X_train.apply(preprocessamento)
ytreino= Y_train

#classificador
classificador_nb_final = clf_pipeline.fit(xtreino,ytreino)

#Salvando modelo
pickle.dump(classificador_nb_final, open('classificador_espam.sav','wb'))
```
